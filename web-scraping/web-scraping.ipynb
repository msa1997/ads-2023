```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

# Setup Selenium driver
driver = webdriver.Chrome()  # Make sure you've installed chromedriver and set it in your PATH

# Target URL
url = "https://bama.ir/car/samand?year=1385"

driver.get(url)

# Script to scroll to the bottom of the page
scroll_pause_time = 10
last_height = driver.execute_script("return document.body.scrollHeight")
for i in range(5):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_pause_time)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# After scrolling, now grab the page source and close the driver
soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.quit()

```

```python

# Find all containers with the cars' information
car_containers = soup.find_all('div', class_='bama-ad-holder')
links = []
for car in car_containers:
    link = car.find("a")
    href = link.get("href") 
    links.append(href)
```

```python
import pandas as pd

# Create a Pandas DataFrame
df = {"Price":[], "Milage":[], "Color":[], "Production":[], "Transmission":[], "Description":[]}
```

```python
import requests
for car in links[:50]:
    html = requests.get("https://bama.ir"+car , headers={'User-Agent': 'Mozilla/5.0'}).content
    # Create a BeautifulSoup object from the HTML content
    new_soup = BeautifulSoup(html, "html.parser")
    if (new_soup.find('div', class_="description-wrapper")):
        if (new_soup.find('div', class_="description-wrapper").find('div', class_="desc")):
            Description = new_soup.find('div', class_="description-wrapper").find('div', class_="desc").find('p').contents
        else:
            Description = ['']
    else:
        Description = ['']
    temp = new_soup.find_all('div', class_="bama-vehicle-detail-with-icon__detail-holder")
    Milage = temp[0].find('p').contents
    Color = temp[4].find('p').contents
    Production = new_soup.find('div', "bama-ad-detail-side-content__top-section-part").find('span', "bama-ad-detail-title__subtitle").contents
    Transmission = temp[2].find('p').contents
    Price = new_soup.find('span', class_="bama-ad-detail-price__price-text").contents

    df["Price"].append(Price[0])
    df["Milage"].append(Milage[0])
    df["Color"].append(Color[0])

    df["Production"].append(Production[0])
    df["Transmission"].append(Transmission[0])
    df["Description"].append(Description[0])
```

```python
df = pd.DataFrame.from_dict(df)
```
